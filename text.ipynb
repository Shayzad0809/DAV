```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
df=pd.read_csv('/content/drive/MyDrive/netflix_titles.csv',usecols=['cast'])
df.head()
```

```python
ndf=df.dropna()
ndf.head()
```

```python
text = " ".join(item for item in ndf['cast'])
print(text)
```

```python
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(background_color="white").generate(text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color="white", max_words=100, max_font_size=300, width=800, height=500, colormap="magma")
wc.generate(text)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
