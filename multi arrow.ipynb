```python
import numpy as np
import matplotlib.pyplot as plt
x_pos = [0, 2, 1]
y_pos = [0, 5, 2]
x_direct = [1, -2, 1]
y_direct = [1, -2, 1]
fig, ax = plt.subplots(figsize=(12, 6))
ax.quiver(x_pos, y_pos, x_direct, y_direct, scale=10.6)
ax.set_title('Quiver plot with multiple arrows')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))
x_pos = x.flatten()
y_pos = y.flatten()
x_direct = np.cos(x)
y_direct = np.sin(y)

fig, ax = plt.subplots(figsize=(10, 10))
ax.quiver(x_pos, y_pos, x_direct, y_direct, scale=20, color='b', alpha=0.5)

ax.set_title('Quiver Plot with Multiple Arrows Forming a Pattern')

plt.show()
```
